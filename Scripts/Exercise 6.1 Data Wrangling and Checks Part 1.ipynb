```python
# 00. Import libraries

import pandas as pd
import numpy as np
import os
```

```python
# 01. Importing primary dataset, causes_of_deaths.csv to create df_causes_of_deaths

df_causes_of_deaths = pd.read_csv(os.path.join(path, 'Data', 'Original Data', 'causes_of_deaths.csv'), index_col = False)
```

```python
# Renaming columns
df_causes_of_deaths.rename(columns={'Country':'country', 'ISO_CODE':'iso_code', 'Year':'year', 'Deaths':'deaths', 'Cause':'cause', 'Male POP':'male_population', 'Female POP':'female_population', 'Total Pop':'total_population', 'GDP':'gdp', 'PCAP':'gdp_per_capita',}, inplace=True)
```

```python
df_causes_of_deaths.head()
```

```python
#Changing the datatype of the variable country, iso_code, cause

df_causes_of_deaths['country'] = df_causes_of_deaths['country'].astype('str')
df_causes_of_deaths['iso_code'] = df_causes_of_deaths['iso_code'].astype('str')
df_causes_of_deaths['cause'] = df_causes_of_deaths['cause'].astype('str')
```

```python
#check for mixed types
for col in df_causes_of_deaths.columns.tolist():
  weird = (df_causes_of_deaths[[col]].applymap(type) != df_causes_of_deaths[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_causes_of_deaths[weird]) > 0:
    print (col)
```

```python
# No mixed data types found
```

```python
#check for missing values
df_causes_of_deaths.isnull().sum()
```

```python
# Create a subset, df_nan_total_population, to view missing values in 'total_population' column
df_nan_total_population = df_causes_of_deaths[df_causes_of_deaths['total_population'].isnull() == True]
```
