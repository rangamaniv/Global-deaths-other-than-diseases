```python
#Import libraries

import pandas as pd
import numpy as np
import os
```

```python
# creating path folder 
path = r'C:\Users\cpaul\Documents\Career Foundry\Causes of deaths other than diseases'
```

```python
# 01. Importing datasets, causes_of_deaths_checked.csv, gdp_data_checked.csv, income_group_data_checked.csv, country_region_data_checked.csv

df_causes_of_deaths = pd.read_csv(os.path.join(path, 'Data', 'Prepared Data', 'causes_of_deaths_checked.csv'), index_col = False)

df_gdp = pd.read_csv(os.path.join(path, 'Data', 'Prepared Data', 'gdp_data_checked.csv'), index_col = False)

df_income_group = pd.read_csv(os.path.join(path, 'Data', 'Prepared Data', 'income_group_data_checked.csv'), encoding = "ISO-8859-1", index_col = False)

df_country_region = pd.read_csv(os.path.join(path, 'Data', 'Prepared Data', 'country_region_data_checked.csv'), encoding = "ISO-8859-1", index_col = False)
```

```python
# Checking dataframe df_causes_of_deaths
df_causes_of_deaths.head()
```

```python
#drop column 'Unnamed: 0'
df_causes_of_deaths = df_causes_of_deaths.drop(columns = ['Unnamed: 0'])
```

```python
# Checking dataframe df_gdp
df_gdp.head()
```

```python
df_gdp = df_gdp.drop(columns = ['iso_code'])
```

```python
df_gdp.head()
```

```python
# Checking dataframe df_income_group
df_income_group.head()
```

```python
#drop column 'Unnamed: 0'
df_income_group = df_income_group.drop(columns = ['Unnamed: 0'])
```

```python
df_income_group.head()
```

```python
# Checking dataframe df_country_region
df_country_region.head()
```

```python
#drop column 'Unnamed: 0'
df_country_region = df_country_region.drop(columns = ['Unnamed: 0'])
```

```python
df_country_region.head()
```

```python
# Drop country column from the other three dataframes

df_gdp = df_gdp.drop(columns = ['country'])
df_income_group = df_income_group.drop(columns = ['country'])
df_country_region = df_country_region.drop(columns = ['country'])
```

```python
# Renaming iso_code column to country_code for uniform names across the datasets

df_causes_of_deaths.rename(columns={'iso_code':'country_code'}, inplace=True)
df_country_region.rename(columns={'iso_code':'country_code'}, inplace=True)
```

# Merge dataframes

```python
#check the shape of df_causes_of_deaths
df_causes_of_deaths.shape
```

```python
#check the shape of df_gdp
df_gdp.shape
```

```python
# Merge the above two dataframes

df_merged_1 = pd.merge(df_causes_of_deaths, df_gdp,  how='left', on=['country_code','year'], indicator = True)
```

```python
df_merged_1.shape
```

```python
df_merged_1.head()
```

```python
#check for full match
df_merged_1['_merge'].value_counts()
```

```python
# Drop the merge indicator column
df_merged_1 = df_merged_1.drop(columns = ['_merge'])
```

```python
df_merged_1.head(20)
```

```python
#check the shape of df_income_group
df_income_group.shape
```

```python
df_income_group.head()
```

```python
# Merge the df_merged_1 dataframe with df_income_group dataframe

df_merged_2 = pd.merge(df_merged_1, df_income_group,  how='left', on=['country_code','year'], indicator = True)
```

```python
df_merged_2.shape
```

```python
df_merged_2.tail(30)
```

```python
# The income_group dataframe has data starting from the year 1987, and so the income_group would be null for yeras from 1980 to 1986.
# However, I would like to subset the data to start from year 1987.
```

```python
#check for full match
df_merged_2['_merge'].value_counts()
```

```python
# Drop the merge indicator column
df_merged_2 = df_merged_2.drop(columns = ['_merge'])
```

```python
df_merged_2.head()
```

```python
# Merge the df_merged_2 dataframe with df_country_regiom dataframe

df_merged_3 = pd.merge(df_merged_2, df_country_region,  how='left', on=['country_code'], indicator = True)
```

```python
df_merged_3.shape
```

```python
df_merged_3.head()
```

```python
#check for full match
df_merged_3['_merge'].value_counts()
```

```python
# Subsetting merged dataset to include data from the year 2007 and after since the project brief stated not to include data more than ten years

df_merged_3_subset =  df_merged_3[df_merged_3['year'] >= 2007]
```

```python
df_merged_3_subset.shape
```

```python
df_merged_3_subset.head()
```

```python
# Drop the merge indicator column
df_merged_3_subset = df_merged_3_subset.drop(columns = ['_merge'])
```

```python
df_merged_3_subset.head()
```

```python
df_merged_3_subset.columns
```

```python
#Export df_merged_3_subset
df_merged_3_subset.to_csv(os.path.join(path, 'Data', 'Prepared Data', 'causes_of_deaths_merged.csv'))
```
